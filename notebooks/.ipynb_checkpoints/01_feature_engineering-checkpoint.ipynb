```python
import pandas as pd
import numpy as np

np.random.seed(42)

# For MVP: Create 25 features from simulated transaction data
# In real: Would transform Kaggle creditcard + synthetic users data

n_users = 50000

features_list = []

for user_id in range(1, n_users + 1):
    # Generate realistic feature values
    
    # TIER 1: RFM (3 features)
    recency_days = np.random.randint(0, 120)
    frequency = np.random.randint(5, 150)
    monetary = np.random.randint(5000, 200000)
    
    # TIER 2: Behavioral (8 features)
    category_diversity = np.random.randint(2, 20)
    
    # Monthly spend (12 features) - distribute total spend across months
    monthly_spends = [np.random.randint(1000, 20000) for _ in range(12)]
    
    avg_transaction_size = monetary / frequency if frequency > 0 else 0
    std_transaction_size = np.random.randint(500, 5000)
    max_transaction = np.random.randint(1000, 50000)
    transaction_trend = np.random.uniform(-0.5, 0.5)  # -0.5 to +0.5
    
    # TIER 3: Demographics (5 features)
    age = np.random.randint(25, 75)
    income_proxy = np.random.randint(30000, 200000)
    card_type_platinum = np.random.randint(0, 2)
    card_type_gold = np.random.randint(0, 2)
    days_as_customer = np.random.randint(30, 1095)
    
    # Create feature row
    feature_row = {
        'user_id': user_id,
        'recency_days': recency_days,
        'frequency': frequency,
        'monetary': monetary,
        'category_diversity': category_diversity,
        **{f'monthly_spend_{i+1}': monthly_spends[i] for i in range(12)},
        'avg_transaction_size': avg_transaction_size,
        'std_transaction_size': std_transaction_size,
        'max_transaction': max_transaction,
        'transaction_trend': transaction_trend,
        'age': age,
        'income_proxy': income_proxy,
        'card_type_platinum': card_type_platinum,
        'card_type_gold': card_type_gold,
        'days_as_customer': days_as_customer
    }
    
    features_list.append(feature_row)

# Create DataFrame
features_df = pd.DataFrame(features_list)

print(f"Features engineered: {features_df.shape}")
print(f"\nColumns ({len(features_df.columns)}):")
for col in features_df.columns:
    print(f"  - {col}")

print(f"\nFeature statistics:")
print(features_df.describe())

# Save
features_df.to_csv('../data/features.csv', index=False)
print(f"\n✓ Features saved to data/features.csv")
print(f"Shape: {features_df.shape} (50K users × 25 features)")
```
