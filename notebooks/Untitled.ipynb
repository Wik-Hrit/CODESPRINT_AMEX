```python
import pandas as pd
import numpy as np

# We'll create the target variable for "Will user redeem Travel Insurance in next 30 days?"

np.random.seed(42)

# Create 50K users
n_users = 50000

user_targets = []
for user_id in range(1, n_users + 1):
    # Simple heuristic: High spenders are more likely to use travel insurance
    total_spend = np.random.randint(1000, 200000)
    transaction_count = np.random.randint(5, 150)
    
    # Target: 1 if high spend + good activity, else 0
    target = 1 if (total_spend > 80000 and transaction_count > 40) else 0
    
    user_targets.append({
        'user_id': user_id,
        'target': target
    })

targets_df = pd.DataFrame(user_targets)

print(f"Target distribution:\n{targets_df['target'].value_counts(normalize=True)}")
print(f"\nShape: {targets_df.shape}")

# Save
targets_df.to_csv('../data/target.csv', index=False)
print(f"\n✓ Target saved to data/target.csv")
```
